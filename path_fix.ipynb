```python

Цей ноутбук автоматично проходить по всіх файлах (TS/TSX/JS/JSX/HTML/CSS/SCSS) і виправляє шляхи типу `images/...` на `/images/...`, якщо слеш на початку відсутній.
</VSCode.Cell>

<VSCode.Cell language="python">import os
from pathlib import Path
import re

root = Path(r"c:\Users\Admin\Desktop\практика\streamvibeReact")

extensions = {'.ts', '.tsx', '.js', '.jsx', '.css', '.scss', '.html'}
files = [p for p in root.rglob('*') if p.suffix in extensions and p.is_file()]
print(f"Знайдено файлів для обробки: {len(files)}")

pattern = re.compile(r'(?<!["\'/])images/')

changes = []
for file in files:
    text = file.read_text(encoding='utf-8')
    new_text = re.sub(r'(?<!["\'/])images/', '/images/', text)
    if new_text != text:
        backup = file.with_suffix(file.suffix + '.bak')
        backup.write_text(text, encoding='utf-8')
        file.write_text(new_text, encoding='utf-8')
        count = text.count('images/') - new_text.count('images/')
        changes.append((file, count))

print(f"Змінено файлів: {len(changes)}")
for f,c in changes[:20]:
    print(f" - {f}: {c} подій")

remaining = []
for file in files:
    text = file.read_text(encoding='utf-8')
    if re.search(r'(?<!["\'/])images/', text):
        remaining.append(file)

print(f"Залишилося проблемних файлів: {len(remaining)}")
for f in remaining[:20]:
    print(f" - {f}")

print('Готово')
```
